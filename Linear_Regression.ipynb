```python
# import necessary modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import random
```

```python
# fix the random seed
random.seed(10)

n_samples = 20 # set the number of observations: training examples

x= np.sort(np.random.rand(n_samples))
true_y = np.cos(1.5 * np.pi * x) # define the underlying function
y = true_y +  np.random.randn(n_samples) * 0.1 # observations: noise has been added to the underlying function value.
```

```python
X_test = np.linspace(0, 1, 100) # The test data: values of the independent variable
Y_test = np.cos(1.5 * np.pi * X_test) # The  test data target from the underlying function
```

```python
X_train = x.reshape(-1, 1)

model_degree_1 = LinearRegression()
model_degree_1.fit(X_train, y)
```

```python
X_test_reshaped = X_test.reshape(-1, 1)
y_pred = model_degree_1.predict(X_test_reshaped)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(X_test, y_pred, label="Model")
plt.plot(X_test, Y_test, label="True function")
plt.scatter(x, y, edgecolor="b", s=20, label="Samples")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim((0, 1))
plt.ylim((-2, 2))
plt.legend(loc="best")
plt.show()
```
