#### Task 1: 
##### Write the code to scatter plot the original data, as of the matrix above, as well as their transformation into a PCA based scatter plot with the covariance matrix being the transformation matrix.

```python
#To plot the original data and the transformed data using PCA with the covariance matrix as the transformation matrix, you can follow these steps using Python and its libraries numpy and matplotlib:


import numpy as np
import matplotlib.pyplot as plt

# Original Data
X = np.array([[-2, 2], [-4, -4], [4, 4], [2, -2], [0, 0]])

# Calculate the mean of the data
mean = np.mean(X, axis=0)

# Center the data
X_centered = X - mean

# Calculate the covariance matrix
cov_matrix = np.cov(X_centered, rowvar=False)

# Perform eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Transform the data using PCA
transformed_data = np.dot(X_centered, eigenvectors)

# Plot original data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c='b', label='Original Data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Original Data')

# Plot PCA transformed data
plt.subplot(1, 2, 2)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c='r', label='PCA Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Transformed Data')
plt.legend()
plt.tight_layout()

plt.show()
```

#### Task 3
##### Modify and run the code for the PCA above, this time, however, by reducing the transformed space by one dimension. Subsequently, study write code to a) display the new data points, b) display the variance ratio being explained, c) display the PCA components as a matrix of eigenvectors, d) display the corresponding singular values.

```python
import numpy as np
from sklearn.decomposition import PCA

# Define the data
X = np.array([[-2, 2], [-4, -4], [4, 4], [2, -2], [0, 0]])

# Perform PCA with reduced dimension
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)

# Display the new data points
print("New Data Points after PCA:")
print(X_pca)

# Display the variance ratio being explained
print("\nVariance Ratio Explained:")
print(pca.explained_variance_ratio_)

# Display the PCA components as a matrix of eigenvectors
print("\nPCA Components (Eigenvectors):")
print(pca.components_)

# Display the corresponding singular values
print("\nSingular Values:")
print(pca.singular_values_)

```

#### Task 4
##### Given the matrix below, decompose the matrix by applying SVD. In particular, you should calculate and display the matrices U, S and V by calculating eigenvalues and eigenvectors rather by simply applying the SVD function.

```python
import numpy as np

# Define the matrix
X = np.array([[-2, 2], [-4, -4], [4, 4], [2, -2], [0, 0]])

# Calculate covariance matrix
covariance_matrix = np.cov(X.T)

# Calculate eigenvectors and eigenvalues of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvectors and eigenvalues
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Calculate matrix U
U = eigenvectors

# Calculate matrix S (diagonal matrix of square roots of eigenvalues)
S = np.diag(np.sqrt(eigenvalues))

# Calculate matrix V (transposed matrix of eigenvectors)
V = np.transpose(eigenvectors)

# Display matrices U, S, and V
print("Matrix U:")
print(U)
print("\nMatrix S:")
print(S)
print("\nMatrix V:")
print(V)
```
